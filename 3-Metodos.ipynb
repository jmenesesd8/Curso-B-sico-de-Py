```python
#Métodos
nombre= str(input("¿Cuál es tu nombre:"))
print (nombre)
print(nombre.upper())#colocar el texto en mayúscula
print(nombre.capitalize())#colocar la primera letra del texto en mayúscula
print(nombre)#si volvemos a llamar nombre, vuelve y aparece en minúscula porque no se ha guardado en la misma variable por tanto:
nombre=nombre.capitalize()
print(nombre)
print(len(nombre))#Devuelve el tamaño de la cadena de texto
nombre=nombre.strip()#elimina los espacios basura de un texto, por ejemplo los espacios
print(nombre)
nombre=nombre.lower()#Coloca en minúscula una cadena de texto
print(nombre)
nombre=nombre.replace('s','x')#reemplaza la letra s por la x
print(nombre)
nombre=nombre[2]#recorrer las letras de un texto
print(nombre)

```
