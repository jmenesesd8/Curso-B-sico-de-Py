```python
#Código profesor de platzi
def num_primo(numero):
  contador=0
  
  for i in range(1,numero+1):
    if i==1 or i==numero:
      continue
    if numero % i==0:
      contador+=1
  if contador==0:
      return True
  else:
     return False




def run():
  numero=int(input("Digite el numero que desee: "))
  if num_primo(numero)==True:
    print("es un numero primo")
  else:
    print("no es un número primo")



if __name__ == '__main__':
    run()
```
