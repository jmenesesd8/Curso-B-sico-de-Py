```python
# De una cadena de txt colocar las letras que están en mayúscula en minúscula y las minúsculas en mayúsculas
def intercambio_de_texto(s):
  concatenado=""
  for letra in s:
    if letra==letra.upper():#letras mayúsculas
        concatenado+=letra.lower()# += símbolo de concatenación y letra.lower es una función relacionada con minúsculas
    else:
      concatenado+=letra.upper()
  print(concatenado)
intercambio_de_texto("jorge ESTEBAN meneses DELGADO")

```
