```python
import random


def generar_contrasena():
    mayusculas = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    minusculas = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
    simbolos = ['!', '#', '$', '&', '/', '(', ')']
    numeros = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']

    caracteres = mayusculas + minusculas + simbolos + numeros #Coloca en una sola lista todos los caracteres concatenandolos
    #print(caracteres)

    contrasena = []#Se crea una lista vacía para ir llenando

    for i in range(15):
        caracter_random = random.choice(caracteres)#Elige un caracter de toda la lista de caracteres
        contrasena.append(caracter_random)# Va Adicionando cada caracter elegido anteriormente a la lista vacia contrasena.Lo hace ciclo por ciclo

    contrasena = "".join(contrasena)#convertir una lista en un string
    return contrasena


def run():
    contrasena = generar_contrasena()
    print('Tu nueva contraseña es: ' + contrasena)


if __name__ == '__main__':
    run()
```
