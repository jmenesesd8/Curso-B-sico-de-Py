```python
#Las tuplas funcionan igual que las lista pero la diferencia es que son inmutables, 
#es decir no le podemos agregar o quitar elementos.Las tuplas son estáticas por tanto, 
#ahorran espacio de memoria y se ejecutan más rápido. Las Tuplas funcionan igual que los strings. A un String no le puede agregar o quitar una letra

tupla=(1,2,3)
print(tupla)
```

```python
for i in tupla:
  print(i)
```
