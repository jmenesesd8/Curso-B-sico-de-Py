```python
def  Conversor(tipo_pesos, valor_dolar):
    pesos=input("¿cuantos pesos " + tipo_pesos +  " tienes?: ")
    pesos=float(pesos)
    dolares=pesos/valor_dolar
    dolares=round(dolares,2)
    dolares=str(dolares)
    print("tiene $"+ dolares+ "dolares")

Menu = """"
           Escoge la opciòn que desees realizar
 
a-Convertir de pesos colombianos a dolares
b-Convertir de pesos Argentino a dolares
c-Convertir de pesos Mexicano a dolares
"""

opcion=str(input(Menu))

if opcion=='a':
    Conversor("colombianos", 3875)
elif opcion=='b':
    Conversor("argentinos", 65)
elif opcion=='c':
    Conversor("mexicanos", 24)
```
